# Analysing Gender


```python
data=pd.DataFrame({'Ratio':[len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])]},
                 index=['Male','Female'])
data.plot.pie(y='Ratio',autopct='%.1f')
```

```python
df.groupby('Gender').size().plot(kind="pie",
                                autopct="%.1f",
                                title='Gender Ratio')
```

```python
df.groupby('Gender').size().plot(kind='bar')
```

```python
df.groupby('Gender').size()
```

```python
gender_purchase = df.groupby('Gender')['Purchase'].sum()
gender_purchase.plot(kind='pie', autopct='%0.1f%%', startangle=90)
plt.title('Total Purchase by Gender')
plt.tight_layout()
plt.show()
```

```python
df.groupby('Gender').mean()['Purchase'].plot(kind = 'pie', autopct = "%0.1f")
```

```python
gender_avg_purchase = df.groupby('Gender')['Purchase'].mean()
gender_avg_purchase.plot(kind='pie', autopct='%0.1f%%', startangle=90)
plt.title('Average Purchase by Gender')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

# Analysing Age & Marital Status


```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12, 6), title = 'Purchase Distribution by Age')
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst , columns = ['Age','Products'])
```

# Combining Age & Marital Status
